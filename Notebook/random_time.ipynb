```python
import random


SEED = 42
random.seed(SEED)

random_list = []

def random_time(u_list):
    
    for _ in range(len(u_list)):
    
        r = random.randint(1419951600, 1427727600)
        random_list.append(r)
    print(len(random_list))
        
    return random_list
```

```python
u_list = list(data['order_id'].unique())
print(len(u_list))
print(data.shape)
```

```python
random_list = random_time(u_list)
```

```python
for o in data['order_id'][:100]:
    print(o)
```

```python
dictionary = dict(zip(u_list, random_list))
print(dictionary)
```

```python
TIME_STAMP_LIST = []
for o in data['order_id']:
    TIME_STAMP_LIST.append(dictionary.get(o))
```

```python
len(TIME_STAMP_LIST)
```

```python
data['TIMESTAMP'] = TIME_STAMP_LIST
```

```python
data.columns = [['ORDER_ID', 'USER_ID', 'ITEM_ID', 'TIMESTAMP']]
```

```python
data = data.astype(int)
```
